```python
from moviepy.editor import VideoFileClip
from IPython.display import Video
```

```python
Video("__temp__.mp4")
```

```python
clip = VideoFileClip("__temp__.mp4")
```

```python
width_of_video1 = clip.w
height_of_video1 = clip.h
print("Width and Height of original video : ", end = " ")
print(str(width_of_video1) + " x ", str(height_of_video1))
```

```python
target_resolution = (1280, 720)
```

```python
clip_resized = clip.resize(0.6)
width_of_video1 = clip_resized.w
height_of_video1 = clip_resized.h
print("Width and Height of original video : ", end = " ")
print(str(width_of_video1) + " x ", str(height_of_video1))
```

```python
clip_resized.write_videofile("__temp__resize.mp4")
```
